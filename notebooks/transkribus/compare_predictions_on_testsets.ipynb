```python
from collections import defaultdict
from pathlib import Path

import pandas as pd

testset_predictions = Path("../../output/test_predictions")


trocr_dfs = [pd.read_csv(e) for e in testset_predictions.glob("**/trocr_*.csv")]
tesseract_dfs = [pd.read_csv(e) for e in testset_predictions.glob("**/tess_*.csv")]
```

```python
for df_list in (trocr_dfs, tesseract_dfs):
    df1 = df_list[0]
    df2 = df_list[1]
    for (urn, page, line), df_ in df1.groupby(["urn", "page", "line"]):
        transcription = df_.transcription.item()
        other_df_transcription = df2[(df2.urn == urn) & (df2.page == page) & (df2.line == line)].transcription.item()
        if transcription != other_df_transcription:
            print(transcription)
            print(other_df_transcription)
            print("---")
```
